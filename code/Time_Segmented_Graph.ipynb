## Imports

The following is all the libraries that must be imported to read, manipulate, and graph the data.

```python
from datascience import *
import numpy as np
import csv

%matplotlib inline
import matplotlib.pyplot as plots
plots.style.use('fivethirtyeight')
```

## Graph

Below is a function that graphs songs for an inputted year and their attributes on a scatter plot.

3 - danceability, 4 - energy, 6 - loudness, 7 - mode, 8 - speechiness, 9 - acousticness, 10 - instrumentalness, 11 - liveness, 12 - valience

```python
from functools import lru_cache as cache
@cache(None)
def attributes(x, y, year):
    d = Table().read_table("../data/cleandata/top_songs_" + str(year) + ".csv")
    d.scatter(x, y, s=100)
    plots.xlim(0,1)
    plots.ylim(0, 1)
    plots.title(year)
```

```python
attributes(3, 4, 2013)
```

## Time Graph

Use `ipywidgets` to combine the graphs with a widget slider.

```python
import ipywidgets as widgets
_ = widgets.interact(attributes, x=3, y=4, year=widgets.IntSlider(min=2012, max=2016, value=2012))
```
